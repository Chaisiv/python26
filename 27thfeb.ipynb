```python
number = 17
is_prime = True
index = 2
while index < number:
    if number % index == 0:
        is_prime = False
    index += 1

if is_prime:
    print("prime number")
else:
    print("not a prime number")
```

```python
number = 151
reverse = 0

while number > 0:
    digit = number % 10
    reverse = reverse * 10  + digit
    number = number // 10

if number == reverse:
    print("pallendrome")
else:
    print("not a pallendrome")
```
