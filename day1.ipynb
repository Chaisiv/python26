```python

percentages = [15 , 80]
for percentage in percentages:
    if 93 <= percentage <= 100:
        print ("A","points = 4") 
    elif 90 <= percentage <= 92:
        print("A-","points = 3.7")
    elif 87 <= percentage <= 89:
        print ("B+","Points = 3.3")
    elif 83 <= percentage <= 86:
        print("B","points = 3")
    elif 80 <= percentage <= 82:
        print("B-","points = 2.7")
    elif 77 <= percentage <= 79:
        print ("c+","Points = 2.3")
    elif 73 <= percentage <= 76 :
        print("C","points = 2")
    else :  
        print ("f")    
```
